```python
df_cleaned = df.drop(['Patient ID', 'Family History', 'Diet', 'Medication Use', 'Income', 'Triglycerides', 'Country', 'Hemisphere'], axis=1)
```

```python
df_cleaned
```

```python
df_cleaned = df_cleaned.rename(columns={'Sex': 'Gender'}) #Renaming the column sex to gender
```

```python
df_cleaned
```

```python
# Spalte aufteilen
df_cleaned[['Blood Pressure systolic', 'Blood Pressure diastolic']] = df_cleaned['Blood Pressure'].str.split('/', expand=True)

# Optional: Werte in Integer umwandeln
df_cleaned['Blood Pressure systolic'] = df_cleaned['Blood Pressure systolic'].astype(int)
df_cleaned['Blood Pressure diastolic'] = df_cleaned['Blood Pressure diastolic'].astype(int)

df_cleaned
```

```python
# Schritt 1: Spaltenliste holen
cols = list(df_cleaned.columns)

# Schritt 2: Zielspalten entfernen, um sie später wieder einzufügen
cols.remove('Blood Pressure systolic')
cols.remove('Blood Pressure diastolic')

# Schritt 3: Neue Spalten an Position 8 und 9 einfügen
cols.insert(4, 'Blood Pressure systolic')
cols.insert(5, 'Blood Pressure diastolic')

# Schritt 4: DataFrame neu anordnen
df_cleaned = df_cleaned[cols]
```

```python
df_cleaned
```

```python
df_cleaned = df_cleaned.drop(columns=['Blood Pressure'])
```

```python
df_cleaned
```

```python
#df_cleaned = df_cleaned[['Age', 'Gender', 'Continent', 'Heart Rate', 'Blood Pressure systolic', 'Blood Pressure diastolic', 'Previous Heart Problems', 'Heart Attack Risk', 'Cholesterol', 'Stress Level', 'Diabetes', 'Smoking', 'Obesity', 'Alcohol Consumption', 'Exercise Hours Per Week', 'Sedentary Hours Per Day', 'Physical Activity Days Per Week', 'Sleep Hours Per Day', 'BMI']]
```

```python
data_gender_female = df_cleaned[df_cleaned['Gender'] == 'Female']
data_gender_male = df_cleaned[df_cleaned['Gender'] == 'Male']
```

```python
data_gender_male
```

```python
data_gender_female
```

```python
#Create new column with age group
bins = [20, 35, 50, 65]
labels = ['Young Adult', 'Adult', 'Senior']


# Neue Spalte 'age_group' hinzufügen
data_gender_female['Age_group'] = pd.cut(data_gender_female['Age'], bins=bins, labels=labels)
data_gender_male['Age_group'] = pd.cut(data_gender_male['Age'], bins=bins, labels=labels)
```

```python
data_gender_female
```

```python
#Sort age group after "age"
cols = list(data_gender_female.columns)
cols.insert(1, cols.pop(cols.index('Age_group')))
data_gender_female = data_gender_female[cols]
```

```python
#Sort age group after "age"
cols = list(data_gender_male.columns)
cols.insert(1, cols.pop(cols.index('Age_group')))
data_gender_male = data_gender_male[cols]
```

```python
#delete all NaN values
data_gender_female_corrected = data_gender_female.dropna(subset=['Age_group'])
data_gender_male_corrected = data_gender_male.dropna(subset=['Age_group'])
```

```python
data_gender_female_corrected
```

```python
data_gender_male_corrected
```

```python
data_gender_male_corrected.isna().sum()
```

```python
#Spalten runden
data_gender_male_corrected[['Exercise Hours Per Week', 'Sedentary Hours Per Day', 'BMI']] = data_gender_male_corrected[['Exercise Hours Per Week', 'Sedentary Hours Per Day', 'BMI']].round(2)
data_gender_female_corrected[['Exercise Hours Per Week', 'Sedentary Hours Per Day', 'BMI']] = data_gender_female_corrected[['Exercise Hours Per Week', 'Sedentary Hours Per Day', 'BMI']].round(2)
```

```python
data_gender_female_corrected.to_csv('data/heartattackprediction/data_heartattackprediction_female_cleaned.csv', index=False)
data_gender_male_corrected.to_csv('data/heartattackprediction/data_heartattackprediction_male_cleaned.csv', index=False)
```
