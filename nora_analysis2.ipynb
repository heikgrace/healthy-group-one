```python
import pandas as pd
import numpy as np
```

```python
data_female = pd.read_csv("data/sleephealthlifestyle/data_gender_female.csv")
```

```python
data_female.head()
```

```python
data_female.info()
```

```python
data_female['Sleep Disorder'].fillna("Unknown", inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(data=data_female, x='Age', bins=10, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(data=data_female, x='BMI Category')
plt.title('BMI Category Distribution')
plt.xlabel('BMI Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Merge Normal and Normal weight

data_female['BMI Category'] = data_female['BMI Category'].replace('Normal', 'Normal Weight')
```

```python
sns.countplot(data=data_female, x='BMI Category')
plt.title('BMI Category Distribution')
plt.xlabel('BMI Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.barplot(data=data_female, x='Age_group', y='Sleep Duration', estimator='mean')
plt.title('Average Sleep Duration by Age Group')
plt.ylabel('Sleep Duration (hours)')
plt.show()
```

```python
sns.boxplot(data=data_female, x='Sleep Disorder', y='Stress Level')
plt.title('Stress Level by Sleep Disorder')
plt.xlabel('Sleep Disorder')
plt.ylabel('Stress Level')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(data_female.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.boxplot(data=data_female, x='BMI Category', y='Physical Activity Level')
plt.title('Physical Activity Level by BMI Category')
plt.xticks(rotation=45)
plt.show()
```

```python
#Stress Level by BMI Category
sns.boxplot(data=data_female, x='BMI Category', y='Stress Level')

plt.xticks(rotation=45)
plt.show()
```

```python
#Sleep Quality by BMI and Sleep Disorder

sns.boxplot(data=data_female, x='BMI Category', y='Quality of Sleep', hue='Sleep Disorder')

plt.xticks(rotation=45)
plt.show()
```

```python
sns.scatterplot(data=data_female, x='Stress Level', y='Quality of Sleep', hue='BMI Category')
plt.title('Quality of Sleep vs Stress Level')
plt.show()
```
