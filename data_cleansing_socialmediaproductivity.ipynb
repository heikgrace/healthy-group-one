```python
df_prepared_columns = df.drop(['social_platform_preference', 'number_of_notifications', 'perceived_productivity_score', 'actual_productivity_score', 'breaks_during_work', 'uses_focus_apps', 'has_digital_wellbeing_enabled', 'days_feeling_burnout_per_month', 'job_satisfaction_score' ], axis=1)
```

```python
#droping rows that have null in stress level because with 30.000 rows, 
# we have enough data to delete 1.904 rows
df_prepared_columns = df_prepared_columns.dropna(subset=['stress_level', 'sleep_hours', 'screen_time_before_sleep' , 'daily_social_media_time'])
```

```python
df_prepared_columns.isna().sum()
```

```python
df_prepared_columns.info()
```

```python
df_prepared_columns
```

```python
df_gender_female = df_prepared_columns[df_prepared_columns['gender'] == 'Female']
df_gender_male = df_prepared_columns[df_prepared_columns['gender'] == 'Male']
```

```python
df_gender_male
```

```python
#exclude 18-22 and 65-79?
#1: 23-30
#2: 31-40
#3: 41-50
#4: 51-65

#Create new column with age group
bins = [20, 35, 50, 65]
labels = ['Young Adult', 'Adult', 'Senior']


# Neue Spalte 'age_group' hinzufügen
df_gender_female['age_group'] = pd.cut(df_gender_female['age'], bins=bins, labels=labels)
df_gender_male['age_group'] = pd.cut(df_gender_male['age'], bins=bins, labels=labels)
```

```python
df_gender_female
```

```python
#Sort age group after "age" in female df
cols = list(df_gender_female.columns)
cols.insert(1, cols.pop(cols.index('age_group')))
df_gender_female = df_gender_female[cols]

#Sort age group after "age" in male df
cols = list(df_gender_male.columns)
cols.insert(1, cols.pop(cols.index('age_group')))
df_gender_male = df_gender_male[cols]
```

```python
df_gender_male.isna().sum()
```

```python
df_gender_female.isna().sum()
```

```python
#delete all NaN values
df_gender_female = df_gender_female.dropna(subset=['age_group'])
df_gender_male = df_gender_male.dropna(subset=['age_group'])
```

```python
df_gender_female = df_gender_female.round(2)
df_gender_male = df_gender_male.round(2)

df_gender_female
```

```python
df_gender_female.to_csv('data/socialmediaproductivity/cleaned_socialmediaproductivity_female.csv', index=False)
df_gender_male.to_csv('data/socialmediaproductivity/cleaned_socialmediaproductivity_male.csv', index=False)
```
