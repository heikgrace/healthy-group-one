```python
data_prepared_columns = data.drop(['survey_code', 'bmi', 'bmi_estimated', 'bmi_scaled', 'occupation', 'healthcare_access', 'insurance', 'family_history', 'gene_marker_flag', 'environmental_risk_score', 'daily_supplement_dosage'], axis=1)
```

```python
#height rounded
data_prepared_columns['height'] = data_prepared_columns['height'].round(2)
#weight rounded
data_prepared_columns['weight'] = data_prepared_columns['weight'].round(2)
#bmi_corrected rounded
data_prepared_columns['bmi_corrected'] = data_prepared_columns['bmi_corrected'].round(2)
#waist_size rounded
data_prepared_columns['waist_size'] = data_prepared_columns['waist_size'].round(2)
#blood_pressure rounded
data_prepared_columns['blood_pressure'] = data_prepared_columns['blood_pressure'].round(2)
#heart rate rounded
data_prepared_columns['heart_rate'] = data_prepared_columns['heart_rate'].round(2)
#cholesterol rounded
data_prepared_columns['cholesterol'] = data_prepared_columns['cholesterol'].round(2)
#insulin rounded
data_prepared_columns['insulin'] = data_prepared_columns['insulin'].round(2)
#glucose rounded
data_prepared_columns['glucose'] = data_prepared_columns['glucose'].round(2)
#sleep_hours rounded
data_prepared_columns['sleep_hours'] = data_prepared_columns['sleep_hours'].round(2)
#work_hours rounded
data_prepared_columns['work_hours'] = data_prepared_columns['work_hours'].round(2)
#physical_activity rounded
data_prepared_columns['physical_activity'] = data_prepared_columns['physical_activity'].round(2)
#daily_steps rounded
data_prepared_columns['daily_steps'] = data_prepared_columns['daily_steps'].round(2)
#calorie_intake rounded
data_prepared_columns['calorie_intake'] = data_prepared_columns['calorie_intake'].round(2)
#sugar_intake rounded
data_prepared_columns['sugar_intake'] = data_prepared_columns['sugar_intake'].round(2)
#water_intake rounded
data_prepared_columns['water_intake'] = data_prepared_columns['water_intake'].round(2)
#screen_time rounded
data_prepared_columns['screen_time'] = data_prepared_columns['screen_time'].round(2)
#income rounded
data_prepared_columns['income'] = data_prepared_columns['income'].round(0).astype('Int64')
```

```python
# Funktion zum Umrechnen
def float_to_time_str(hours):
    h = int(hours)
    m = round((hours - h) * 60)
    return f"{h:02d}:{m:02d}"

# Neue Spalte mit formatierten Zeiten
#data_prepared_columns['sleep_hours'] = data_prepared_columns['sleep_hours'].apply(float_to_time_str)
data_prepared_columns['work_hours'] = data_prepared_columns['work_hours'].apply(float_to_time_str)
data_prepared_columns['screen_time'] = data_prepared_columns['screen_time'].apply(float_to_time_str)
```

```python
#Group by gender

data_gender_female = data_prepared_columns[data_prepared_columns['gender'] == 'Female']
data_gender_male = data_prepared_columns[data_prepared_columns['gender'] == 'Male']
```

```python
data_gender_male
```

```python
#exclude 18-22 and 65-79?
#1: 23-30
#2: 31-40
#3: 41-50
#4: 51-65

#Create new column with age group
bins = [20, 35, 50, 65]
labels = ['Young Adult', 'Adult', 'Senior']


# Neue Spalte 'age_group' hinzufügen
data_gender_female['age_group'] = pd.cut(data_gender_female['age'], bins=bins, labels=labels)
data_gender_male['age_group'] = pd.cut(data_gender_male['age'], bins=bins, labels=labels)
```

```python
data_gender_female
```

```python
#Sort age group after "age"
cols = list(data_gender_female.columns)
cols.insert(1, cols.pop(cols.index('age_group')))
data_gender_female = data_gender_female[cols]
```

```python
#Sort age group after "age"
cols = list(data_gender_male.columns)
cols.insert(1, cols.pop(cols.index('age_group')))
data_gender_male = data_gender_male[cols]
```

```python
#delete all NaN values
data_gender_female_corrected = data_gender_female.dropna(subset=['age_group'])
data_gender_male_corrected = data_gender_male.dropna(subset=['age_group'])
```
