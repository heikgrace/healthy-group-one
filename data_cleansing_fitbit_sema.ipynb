```python
df_prepared_columns = df.drop(['Unnamed: 0', 'activityType', 'date', 'nightly_temperature', 'nremhr',
 'spo2', 'full_sleep_breathing_rate', 'exertion_points_percentage',
 'responsiveness_points_percentage', 'daily_temperature_variation',
 'badgeType', 'filteredDemographicVO2Max', 'mindfulness_session', 
 'scl_avg', 'resting_hr', 'minutesToFallAsleep', 'minutesAsleep', 'minutesAwake',
  'minutesAfterWakeup', 'sleep_deep_ratio', 'sleep_wake_ratio', 'sleep_light_ratio',
  'sleep_rem_ratio', 'minutes_in_default_zone_1', 'minutes_below_default_zone_1', 
  'minutes_in_default_zone_2', 'minutes_in_default_zone_3', 'step_goal', 'min_goal',
   'max_goal', 'step_goal_label', 'ALERT', 'HAPPY', 'NEUTRAL', 'RESTED/RELAXED',
   'SAD', 'TENSE/ANXIOUS', 'TIRED', 'ENTERTAINMENT', 'GYM', 'HOME', 'HOME_OFFICE',
   'OTHER', 'OUTDOORS', 'TRANSIT', 'WORK/SCHOOL' ], axis=1)
```

```python
df_prepared_columns.info()
```

```python
cols = list(df_prepared_columns.columns)
cols.insert(0, cols.pop(cols.index('age')))
cols.insert(1, cols.pop(cols.index('gender')))
cols.insert(2, cols.pop(cols.index('bmi')))
df_prepared_columns = df_prepared_columns[cols]
```

```python
df_prepared_columns['bmi'].tolist()
```

```python
df_prepared_columns['age'].isna().sum()
```

```python
df_prepared_columns = df_prepared_columns.dropna(subset=['age'])
```

```python
df_prepared_columns['age'].tolist()
```

```python
df_prepared_columns['age_group'] = df_prepared_columns['age'].apply(
    lambda x: 'Young Adults' if x == '<30' else 'Adults'
)
```

```python
print(df_prepared_columns[['age', 'age_group']].head())
```

```python
df_prepared_columns
```

```python
cols = list(df_prepared_columns.columns)
cols.insert(1, cols.pop(cols.index('age_group')))
df_prepared_columns = df_prepared_columns[cols]
```

```python
df_prepared_columns
```

```python
df_prepared_columns.round(2)
df_prepared_columns['gender'] = df_prepared_columns['gender'].str.lower()
df_prepared_columns['age_group'] = df_prepared_columns['age_group'].str.lower()
df_prepared_columns
```

```python
df_prepared_columns['stress_score'].describe()
```

```python
df_prepared_columns = df_prepared_columns.dropna(subset=['stress_score', 'rmssd', 'distance', 'bpm', 'sleep_duration', 'sleep_efficiency', 'steps']).reset_index(drop=True)
```

```python
amountOfUniqueIds = df_prepared_columns['id'].nunique()
print(amountOfUniqueIds)
```

```python
# Mapping erstellen: alte ID → neue Nummer (beginnend bei 1)
id_mapping = {old_id: new_id for new_id, old_id in enumerate(df_prepared_columns['id'].unique(), start=1)}

# Alte ID-Spalte ersetzen durch numerische ID
df_prepared_columns['id'] = df_prepared_columns['id'].map(id_mapping)

df_prepared_columns['sleep_duration'] = df_prepared_columns['sleep_duration'] / 3600000


# 'id' als erste Spalte setzen
cols = ['id'] + [col for col in df_prepared_columns.columns if col != 'id']
df_prepared_columns = df_prepared_columns[cols]

```
