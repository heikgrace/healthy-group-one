## What is the individual's overall clinical health status based on these vitals (scaled 0-100)?

### Vitals considered:
- BMI
- Blood Pressure
- Cholesterol
- Heart Rate
- Glucose
- Insulin

### Scale:
- Values are scaled from 0 to 100
- 100 indicates excellent health

---

**Note:**  
The overall clinical health status is calculated by aggregating the scaled values of the above vitals into a composite score ranging from 0 (poor health) to 100 (excellent health).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: Wenn nicht schon geschehen
# df_female = pd.read_csv("...")  # dein DataFrame laden


correlation_vars = [
    'blood_pressure',
    'age', 'bmi_corrected', 'waist_size',
    'physical_activity', 'sleep_hours', 'sleep_quality',
    'stress_level', 'alcohol_consumption', 'smoking_level',
    'calorie_intake', 'sugar_intake', 'heart_rate',
    'cholesterol', 'glucose', 'insulin',
    'water_intake', 'screen_time', 'electrolyte_level'
]

# DataFrame nur mit diesen Spalten
df_corr = df_female[correlation_vars].dropna()  # NaN entfernen für saubere Korrelation
```

```python
df_corr.info
```

```python
# Copy zum Bearbeiten
df = df_female.copy()

# 🟢 sleep_quality: Ordinalskala
sleep_map = {
    'Poor': 1,
    'Fair': 2,
    'Good': 3,
    'Excellent': 4
}
df['sleep_quality'] = df['sleep_quality'].map(sleep_map)

age_group_map = {
    'Teen': 1,
    'Young Adult': 2,
    'Adult': 3,
    'Middle-aged': 4,
    'Senior': 5,
    'Elderly': 6
}

df['age_group'] = df['age_group'].map(age_group_map)

gender_map = {
    'Female': 0,
    'Male': 1,
    'Other': 2  # Falls vorhanden
}

df['gender'] = df['gender'].map(gender_map)

# 🟢 alcohol_consumption: Ordinalskala
alcohol_map = {
    'Never': 0,
    'Occasionally': 1,
    'Regularly': 2,
    'Heavily': 3
}
df['alcohol_consumption'] = df['alcohol_consumption'].map(alcohol_map)

# 🟢 smoking_level: Ordinalskala
smoking_map = {
    'Non-smoker': 0,
    'Light': 1,
    'Heavy': 2
}
df['smoking_level'] = df['smoking_level'].map(smoking_map)

# 🟢 screen_time: HH:MM → Dezimalstunden
# Falls screen_time als String wie '04:07' vorliegt:
df['screen_time'] = pd.to_timedelta(df['screen_time'] + ':00').dt.total_seconds() / 3600.0

```

```python
df
```

```python
corr_matrix = df.corr()
```

```python
bp_corr = corr_matrix['blood_pressure'].sort_values(ascending=False)
print(bp_corr)
```

```python
plt.figure(figsize=(10, 6))
bp_corr.drop('blood_pressure').plot(kind='barh', color='teal')
plt.title('Korrelation mit Blutdruck (blood_pressure)')
plt.xlabel('Korrelationskoeffizient')
plt.axvline(0, color='gray', linestyle='--')
plt.grid(True, axis='x')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(df_corr.corr(), annot=True, fmt=".2f", cmap='coolwarm', center=0)
plt.title("Korrelationsmatrix – ausgewählte Variablen")
plt.tight_layout()
plt.show()
```
