```python
# Spalte aufteilen
data_prepared_columns[['Blood Pressure systolic', 'Blood Pressure diastolic']] = data_prepared_columns['Blood Pressure'].str.split('/', expand=True)

# Optional: Werte in Integer umwandeln
data_prepared_columns['Blood Pressure systolic'] = data_prepared_columns['Blood Pressure systolic'].astype(int)
data_prepared_columns['Blood Pressure diastolic'] = data_prepared_columns['Blood Pressure diastolic'].astype(int)

data_prepared_columns
```

```python
# Schritt 1: Spaltenliste holen
cols = list(data_prepared_columns.columns)

# Schritt 2: Zielspalten entfernen, um sie spÃ¤ter wieder einzufÃ¼gen
cols.remove('Blood Pressure systolic')
cols.remove('Blood Pressure diastolic')

# Schritt 3: Neue Spalten an Position 8 und 9 einfÃ¼gen
cols.insert(8, 'Blood Pressure systolic')
cols.insert(9, 'Blood Pressure diastolic')

# Schritt 4: DataFrame neu anordnen
data_prepared_columns = data_prepared_columns[cols]
```

```python
data_prepared_columns = data_prepared_columns.drop(columns=['Blood Pressure'])
```

```python
data_prepared_columns = data_prepared_columns.fillna('')
```

```python
data_gender_female = data_prepared_columns[data_prepared_columns['Gender'] == 'Female']
data_gender_male = data_prepared_columns[data_prepared_columns['Gender'] == 'Male']
```

```python
data_gender_male
```

```python
#exclude 18-22 and 65-79?
#1: 23-30
#2: 31-40
#3: 41-50
#4: 51-65

#Create new column with age group
bins = [20, 35, 50, 65]
labels = ['Young Adult', 'Adult', 'Senior']


# Neue Spalte 'age_group' hinzufÃ¼gen
data_gender_female['Age_group'] = pd.cut(data_gender_female['Age'], bins=bins, labels=labels)
data_gender_male['Age_group'] = pd.cut(data_gender_male['Age'], bins=bins, labels=labels)
```

```python
data_gender_female
```

```python
#Sort age group after "age"
cols = list(data_gender_female.columns)
cols.insert(1, cols.pop(cols.index('Age_group')))
data_gender_female = data_gender_female[cols]
```

```python
#Sort age group after "age"
cols = list(data_gender_male.columns)
cols.insert(1, cols.pop(cols.index('Age_group')))
data_gender_male = data_gender_male[cols]
```

```python
#delete all NaN values
data_gender_female_corrected = data_gender_female.dropna(subset=['Age_group'])
data_gender_male_corrected = data_gender_male.dropna(subset=['Age_group'])
```

```python
data_gender_female_corrected['BMI Category'] = data_gender_female_corrected['BMI Category'].replace('Normal', 'Normal Weight')
data_gender_male_corrected['BMI Category'] = data_gender_male_corrected['BMI Category'].replace('Normal', 'Normal Weight')
```

```python
data_gender_female_corrected.to_csv('data/sleephealthlifestyle/data_gender_female.csv', index=False)
data_gender_male_corrected.to_csv('data/sleephealthlifestyle/data_gender_male.csv', index=False)
```

```python
combined_df = pd.concat([data_gender_female_corrected, data_gender_male_corrected], ignore_index=True)
combined_df.to_csv('data/sleephealthlifestyle/data_gender_combined.csv', index=False)
```
